````python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Define SMAPE function
def smape(y_true, y_pred):
    return 100 * np.mean(2 * np.abs(y_pred - y_true) / (np.abs(y_pred) + np.abs(y_true) + 1e-8))

# Load dataset
df = pd.read_csv("/content/holidays.csv")
df['datum'] = pd.to_datetime(df['datum'])
df = df.set_index('datum')

# Drug columns
drug_cols = ['M01AB', 'M01AE', 'N02BA', 'N02BE', 'N05B', 'N05C', 'R03', 'R06']

# Exogenous features
exog_cols = ['public_holiday', 'new_year', 'statehood_day', 'labor_day', 'armistice_day',
             'orthodox_christmas', 'orthodox_easter_sunday', 'orthodox_easter_monday',
             'orthodox_good_friday', 'orthodox_holy_saturday', 'cold_flu_season', 'pollen_season']
exog = df[exog_cols].fillna(0)

# Utility functions
def create_lags(series, lags=14):
    return pd.DataFrame({f'lag_{i}': series.shift(i) for i in range(1, lags+1)})

def create_rolling(series):
    return pd.DataFrame({
        'roll_mean_3': series.rolling(3).mean(),
        'roll_mean_7': series.rolling(7).mean(),
        'roll_mean_14': series.rolling(14).mean(),
        'roll_std_7': series.rolling(7).std(),
        'roll_std_14': series.rolling(14).std()
    })

def create_features(series):
    lagged = create_lags(series)
    rolling = create_rolling(series)
    seasonal = pd.DataFrame({
        'dayofweek': series.index.dayofweek,
        'month': series.index.month,
        'weekofyear': series.index.isocalendar().week,
        'quarter': series.index.quarter,
        'is_non_zero': (series > 0).astype(int)
    }, index=series.index)
    return pd.concat([lagged, rolling, seasonal, exog], axis=1).dropna()

def prepare_future_exog(df, exog_cols, start_date, periods=5):
    future_dates = pd.date_range(start=start_date, periods=periods, freq='D')
    future_exog = pd.DataFrame(0, index=future_dates, columns=exog_cols)
    last_known = df[exog_cols].iloc[-1]
    for col in exog_cols:
        future_exog[col] = last_known[col]
    return future_exog

# Forecasting function with ARIMA
def forecast_drug(drug):
    print(f"\nðŸ”¹ Forecasting {drug} with ARIMA...")

    y_orig = df[drug].clip(lower=0)  # Original values
    zero_ratio = (y_orig == 0).mean()
    use_log = zero_ratio < 0.5
    if use_log:
        y = np.log1p(y_orig)
    else:
        y = y_orig.copy()

    # Create features
    X = create_features(y_orig)
    y_supervised = y.loc[X.index]
    exog_supervised = X[exog_cols]

    # Time series split
    from sklearn.model_selection import TimeSeriesSplit
    tscv = TimeSeriesSplit(n_splits=5)
    results = {'RMSE': [], 'MAE': [], 'SMAPE': [], 'MSE': [], 'R2': [], 'Order': []}
    models = []

    for fold, (train_idx, test_idx) in enumerate(tscv.split(X), 1):
        print(f"  - Fold {fold}/5")
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y_supervised.iloc[train_idx], y_supervised.iloc[test_idx]
        exog_train, exog_test = exog_supervised.iloc[train_idx], exog_supervised.iloc[test_idx]

        # Fit ARIMA model with fixed order (1,1,1)
        try:
            model = ARIMA(y_train, exog=exog_train, order=(1, 1, 1)).fit()
            models.append(model)

            # Predictions
            yhat_test = model.forecast(steps=len(test_idx), exog=exog_test)
            if use_log:
                yhat_test = np.expm1(yhat_test)
                y_test_orig = np.expm1(y_test)
            else:
                y_test_orig = y_test

            # Clip negative values
            yhat_test = np.clip(yhat_test, 0, None)

            # Calculate metrics
            mse = mean_squared_error(y_test_orig, yhat_test)
            results['MSE'].append(mse)
            results['RMSE'].append(np.sqrt(mse))
            results['MAE'].append(mean_absolute_error(y_test_orig, yhat_test))
            results['SMAPE'].append(smape(y_test_orig, yhat_test))
            results['R2'].append(r2_score(y_test_orig, yhat_test))
            results['Order'].append((1, 1, 1))
        except Exception as e:
            print(f"  - Error in fold {fold}: {e}")
            results['MSE'].append(np.nan)
            results['RMSE'].append(np.nan)
            results['MAE'].append(np.nan)
            results['SMAPE'].append(np.nan)
            results['R2'].append(np.nan)
            results['Order'].append((1, 1, 1))

    # Average metrics
    avg_results = {k: np.nanmean(v) for k, v in results.items() if k != 'Order'}
    avg_results['Order'] = (1, 1, 1)  # Fixed order
    print(f"  -> Avg Test Metrics: RMSE={avg_results['RMSE']:.2f}, MAE={avg_results['MAE']:.2f}, "
          f"SMAPE={avg_results['SMAPE']:.2f}%, MSE={avg_results['MSE']:.2f}, R2={avg_results['R2']:.2f}, "
          f"Order={avg_results['Order']}")

    # Forecast next 5 days
    future_exog = prepare_future_exog(df, exog_cols, df.index[-1] + pd.Timedelta(days=1), periods=5)
    last_vals = y_orig[-14:].tolist()
    future_forecast = []

    # Refit model on full data for forecasting
    try:
        final_model = ARIMA(y_supervised, exog=exog_supervised, order=(1, 1, 1)).fit()

        for i in range(5):
            lagged_vals = pd.DataFrame({f'lag_{j+1}': [last_vals[-14+j]] for j in range(14)})
            rolling_vals = pd.DataFrame({
                'roll_mean_3': [np.mean(last_vals[-3:])],
                'roll_mean_7': [np.mean(last_vals[-7:])],
                'roll_mean_14': [np.mean(last_vals)],
                'roll_std_7': [np.std(last_vals[-7:])],
                'roll_std_14': [np.std(last_vals)]
            })
            seasonal_vals = pd.DataFrame({
                'dayofweek': [future_exog.index[i].dayofweek],
                'month': [future_exog.index[i].month],
                'weekofyear': [future_exog.index[i].isocalendar().week],
                'quarter': [future_exog.index[i].quarter],
                'is_non_zero': [int(last_vals[-1] > 0)]
            })
            X_future = pd.concat([lagged_vals, rolling_vals, seasonal_vals, future_exog.iloc[[i]]], axis=1)
            X_future = X_future.reindex(columns=X.columns, fill_value=0)
            pred = final_model.forecast(steps=1, exog=X_future[exog_cols])[0]
            pred = max(0, np.expm1(pred) if use_log else pred)  # Clip negative
            future_forecast.append(pred)
            last_vals.append(pred)
            # Update model with new observation
            final_model = ARIMA(
                pd.concat([pd.Series(final_model.data.endog), pd.Series([pred])]),
                exog=pd.concat([pd.DataFrame(final_model.data.exog), X_future[exog_cols]]),
                order=(1, 1, 1)
            ).fit()
    except Exception as e:
        print(f"Error in final forecast for {drug}: {e}")
        future_forecast = [np.nan] * 5

    # Plot last fold
    plt.figure(figsize=(12, 5))
    plt.plot(y_test_orig.index, y_test_orig.values, label='Actual')
    plt.plot(y_test_orig.index, yhat_test, label='Predicted', color='red')
    plt.title(f"ARIMA Forecast for {drug} (Last Fold)")
    plt.legend()
    plt.show()

    # Print next 5 days
    print("\nNext 5-Day Forecast:")
    for i, date in enumerate(future_exog.index):
        print(f"{date.date()}: {future_forecast[i]:.2f}")

    return drug, avg_results, future_forecast, future_exog.index

# Run for all drugs
results = {}
future_forecasts = {}

for drug in drug_cols:
    drug_name, metrics, forecast, dates = forecast_drug(drug)
    results[drug_name] = metrics
    future_forecasts[drug_name] = pd.Series(forecast, index=dates)

# Display summary metrics
print("\nðŸ“Š ARIMA Forecast Accuracy Summary:")
summary_df = pd.DataFrame(results).T
print(summary_df[['MAE', 'MSE', 'RMSE', 'SMAPE', 'R2', 'Order']])
````

````python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from pmdarima import auto_arima

# Load dataset
file_path = "/content/holidays.csv"
data = pd.read_csv(file_path)

# Convert 'datum' to datetime and set as index
data['datum'] = pd.to_datetime(data['datum'])
data.set_index('datum', inplace=True)

# List of drug columns
drug_columns = ['M01AB', 'M01AE', 'N02BA', 'N02BE', 'N05B', 'N05C', 'R03', 'R06']

# Dictionary to store results
results = {}

for drug in drug_columns:
    print(f"\nðŸ”¹ Forecasting for {drug}...")

    # Extract time series
    ts = data[drug].fillna(method='ffill')  # fill missing values

    # Train-test split (last 10 for testing)
    train = ts[:-10]
    test = ts[-10:]

    # Fit Auto ARIMA
    model = auto_arima(
        train,
        start_p=0, max_p=3,
        start_q=0, max_q=3,
        d=None, max_d=2,
        seasonal=False,  # change to True if you suspect seasonality
        trace=False,
        error_action='ignore',
        suppress_warnings=True,
        stepwise=True
    )

    # Forecast
    forecast, conf_int = model.predict(n_periods=10, return_conf_int=True)

    # Evaluation
    mse = mean_squared_error(test, forecast)
    mae = mean_absolute_error(test, forecast)
    r2 = r2_score(test, forecast)

    results[drug] = {"MSE": mse, "MAE": mae, "R2": r2}

    print(f"{drug} â†’ MSE={mse:.4f}, MAE={mae:.4f}, RÂ²={r2:.4f}")

    # Plot
    plt.figure(figsize=(12, 6))
    plt.plot(train.index, train, label="Training Data")
    plt.plot(test.index, test, label="Test Data", color="blue")
    plt.plot(test.index, forecast, label="Forecast", color="red")
    plt.fill_between(test.index, conf_int[:, 0], conf_int[:, 1],
                     color="red", alpha=0.1, label="95% Confidence Interval")
    plt.title(f"Auto ARIMA Forecast for {drug}")
    plt.xlabel("Date")
    plt.ylabel("Value")
    plt.legend()
    plt.show()

# Show final performance table
results_df = pd.DataFrame(results).T
print("\nðŸ“Š Forecasting Performance (last 10 points):")
print(results_df)
````

````python
```python
# Install required packages


import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Define SMAPE function
def smape(y_true, y_pred):
    return 100 * np.mean(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred) + 1e-10))

# Upload dataset in Colab
from google.colab import files
uploaded = files.upload()  # Upload 'holidays.csv'
data = pd.read_csv('/content/holidays.csv')

# Preprocess data
data['datum'] = pd.to_datetime(data['datum'])
data.set_index('datum', inplace=True)

# Define target and exogenous variables
target_column = 'N05B'
exog_columns = ['public_holiday', 'dow', 'weekofyear', 'month', 'cold_flu_season', 'pollen_season']
time_series = data[target_column]
exog_data = data[exog_columns]

# Scale the data
scaler_y = MinMaxScaler()
scaler_exog = MinMaxScaler()
y_scaled = scaler_y.fit_transform(time_series.values.reshape(-1, 1))
exog_scaled = scaler_exog.fit_transform(exog_data)

# Create sequences for LSTM (7-day lookback)
def create_sequences(y, exog, time_steps=7):
    X, X_exog, y_seq = [], [], []
    for i in range(len(y) - time_steps):
        X.append(y[i:i + time_steps])
        X_exog.append(exog[i + time_steps])
        y_seq.append(y[i + time_steps])
    return np.array(X), np.array(X_exog), np.array(y_seq)

time_steps = 7
X, X_exog, y_seq = create_sequences(y_scaled, exog_scaled, time_steps)

# Split into train and test sets (last 10 points for testing)
train_size = len(X) - 10
X_train, X_test = X[:train_size], X[train_size:]
X_exog_train, X_exog_test = X_exog[:train_size], X_exog[train_size:]
y_train, y_test = y_seq[:train_size], y_seq[train_size:]

# Build LSTM model
model = Sequential([
    LSTM(50, activation='relu', input_shape=(time_steps, 1), return_sequences=True),
    Dropout(0.2),
    LSTM(50, activation='relu'),
    Dropout(0.2),
    Dense(len(exog_columns)),  # For exogenous variables
    Dense(1)  # Output layer
])

# Compile model
model.compile(optimizer='adam', loss='mse')

# Reshape input to [samples, time_steps, features]
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Combine LSTM input with exogenous variables
X_train_combined = [X_train, X_exog_train]
X_test_combined = [X_test, X_exog_test]

# Train model
history = model.fit(
    X_train_combined, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.1,
    verbose=1
)

# Forecast for test period
y_pred_scaled = model.predict(X_test_combined)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_test_orig = scaler_y.inverse_transform(y_test)

# Calculate metrics
mse = mean_squared_error(y_test_orig, y_pred)
mae = mean_absolute_error(y_test_orig, y_pred)
rmse = np.sqrt(mse)
smape_val = smape(y_test_orig, y_pred)
r2 = r2_score(y_test_orig, y_pred)

# Print metrics
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"SMAPE: {smape_val:.4f}%")
print(f"RÂ²: {r2:.4f}")

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(data.index[-10:], y_test_orig, label='Test Data', color='blue')
plt.plot(data.index[-10:], y_pred, label='Forecast', color='red')
plt.title('LSTM Forecast for N05B')
plt.xlabel('Date')
plt.ylabel('N05B Value')
plt.legend()
plt.show()

# Plot training history
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
````
